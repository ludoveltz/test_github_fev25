3. Température et Biais

Impact de la Température sur les Biais :

Basse Température (0.2)
- Amplifie les biais existants
- Renforce les stéréotypes dominants
- Moins de diversité dans les outputs

Haute Température (1.5)
- Peut diluer certains biais
- Introduit de la variabilité
- Risque de contenus inappropriés

Exemple Pratique pour une agence vidéo :
- Tâche : Génération de descriptions visuelles
- Basse Température : Peut favoriser systématiquement certains styles/genres
- Haute Température : Peut introduire plus de diversité mais moins de contrôle
- Solution : Température adaptative (0.7-0.9) avec filtres de contenu
