```python
# Imports
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
import gradio as gr

# Initialisation du modèle de dialogue
model_name = "bigscience/bloom-3b"  # Meilleur support du français et plus sophistiqué
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Contexte luxe pour le prompt
LUXURY_CONTEXT = """
Vous êtes l'assistant IA de Solweig & Izar, cabinet de conseil spécialisé dans le luxe français.
Expertise : création de visuels luxe, design haute couture, joaillerie et accessoires premium.
Clients : Maisons de luxe comme Loewe, L'Oréal, Cartier, Gucci.
Style : élégant, raffiné, précis, avec une excellente connaissance des codes du luxe français.
"""

# Initialisation des listes pour l'historique
message_list = []
response_list = []

def mini_chatbot(message, history):
    # Création du prompt enrichi
    prompt = f"{LUXURY_CONTEXT}\nQuestion : {message}\nRéponse :"

    # Génération de la réponse avec paramètres optimisés
    response = generator(
        prompt,
        max_length=200,          # Réponses plus détaillées
        num_return_sequences=1,
        temperature=0.7,         # Plus créatif tout en restant cohérent
        top_p=0.9,              # Meilleure diversité des réponses
        top_k=50,               # Maintient la pertinence
        repetition_penalty=1.2,  # Évite les répétitions
        pad_token_id=tokenizer.eos_token_id
    )[0]['generated_text']

    # Nettoyage de la réponse
    response = response.split("Réponse :")[1].strip()

    # Mise à jour de l'historique
    message_list.append(message)
    response_list.append(response)

    return response

# Création de l'interface Gradio avec style luxe
demo_chatbot = gr.ChatInterface(
    mini_chatbot,
    title="🎯 L'Expert Luxe - Solweig & Izar",
    description="""
    Votre conseiller en excellence pour le luxe français.
    Spécialisé en création visuelle haut de gamme et innovation IA.
    Une expertise unique au service des plus grandes maisons.
    """,
    theme="soft",  # Theme plus élégant
    examples=[
        ["Comment créer une campagne visuelle pour une collection haute couture ?"],
        ["Quelles sont les dernières tendances en joaillerie de luxe ?"],
        ["Comment intégrer l'IA dans le processus créatif d'une maison de luxe ?"]
    ]
)

# Lancement de l'interface
demo_chatbot.launch(share=True)
```
