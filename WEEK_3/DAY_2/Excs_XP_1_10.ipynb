```python
import numpy as np

print("EXERCICE 1 : OPÉRATIONS MATRICIELLES")

# 1. Création d'une matrice 3x3
matrice = np.array([
    [4, 2, -1],
    [1, 5, 3],
    [2, -2, 6]
])
print("\nMatrice 3x3 :")
print(matrice)

# 2. Calcul du déterminant
determinant = np.linalg.det(matrice)
print(f"\nDéterminant de la matrice : {determinant:.2f}")

# 3. Calcul de l'inverse
inverse = np.linalg.inv(matrice)
print("\nInverse de la matrice :")
# Arrondir chaque élément de la matrice inverse à 2 décimales
inverse_arrondie = np.round(inverse, decimals=2)
print(inverse_arrondie)

print("\n" + "="*50 + "\n")  # Séparateur
```

```python
# EXERCICE 2 : ANALYSE STATISTIQUE
print("EXERCICE 2 : ANALYSE STATISTIQUE")

# 1. Génération d'un tableau de 50 nombres aléatoires
np.random.seed(42)  # Pour la reproductibilité
donnees = np.random.normal(loc=50, scale=10, size=50)
print("\nTableau de données généré (arrondi à 2 décimales) :")
print(np.round(donnees, decimals=2))

# 2. Calcul des mesures statistiques
moyenne = np.mean(donnees)
mediane = np.median(donnees)
ecart_type = np.std(donnees)

print(f"\nMoyenne : {moyenne:.2f}")
print(f"Médiane : {mediane:.2f}")
print(f"Écart-type : {ecart_type:.2f}")
```

```python
print("EXERCICE 3 : MANIPULATION DES DATES")

# 1. Création d'un tableau de dates pour janvier 2023
dates = np.arange('2023-01-01', '2023-02-01', dtype='datetime64[D]')
print("\nDates originales :")
print(dates)

# 2. Conversion des dates dans un format YYYY/MM/DD
dates_formatees = np.datetime_as_string(dates, unit='D')
dates_formatees = [date.replace('-', '/') for date in dates_formatees]

print("\nDates reformatées (YYYY/MM/DD) :")
for date in dates_formatees:
    print(date)

print(f"\nNombre total de jours : {len(dates)}")
```

```python
# Imports nécessaires
import numpy as np
import pandas as pd

print("EXERCICE 4 : MANIPULATION DE DONNÉES AVEC NUMPY ET PANDAS")

# 1. Création d'un DataFrame avec des nombres aléatoires
np.random.seed(42)  # Pour la reproductibilité
data = {
    'A': np.random.randint(0, 100, 10),
    'B': np.random.randint(0, 100, 10),
    'C': np.random.randint(0, 100, 10)
}

df = pd.DataFrame(data)
print("\nDataFrame original :")
print(df)

# 2. Sélection conditionnelle
print("\nValeurs de la colonne A supérieures à 50 :")
selection = df[df['A'] > 50]
print(selection)

# 3. Fonctions d'agrégation
print("\nStatistiques d'agrégation :")
print(f"Somme de chaque colonne :")
print(df.sum().round(2))

print(f"\nMoyenne de chaque colonne :")
print(df.mean().round(2))
```

```python
# Imports nécessaires
import matplotlib.pyplot as plt

print("EXERCICE 5 : REPRÉSENTATION D'IMAGES")

# 1. Création d'une image en niveaux de gris 5x5
image = np.array([
    [0, 50, 100, 50, 0],    # Première ligne
    [50, 100, 200, 100, 50], # Deuxième ligne
    [100, 200, 255, 200, 100], # Ligne centrale
    [50, 100, 200, 100, 50], # Quatrième ligne
    [0, 50, 100, 50, 0]     # Dernière ligne
])

print("\nReprésentation numérique de l'image (matrice 5x5) :")
print(image)

# 2. Affichage de l'image
plt.figure(figsize=(5, 5))
plt.imshow(image, cmap='gray')
plt.colorbar(label='Intensité')
plt.title('Image en niveaux de gris 5x5')
plt.axis('on')  # Affiche les axes pour voir les dimensions
plt.show()

print("\nExplication :")
print("- Les valeurs vont de 0 (noir) à 255 (blanc)")
print("- Chaque nombre représente l'intensité d'un pixel")
print("- La matrice 5x5 crée un motif simple en dégradé")

```

```python
from scipy import stats

print("EXERCICE 6 : TEST D'HYPOTHÈSE")

# 1. Création des données
np.random.seed(42)  # Pour la reproductibilité
productivity_before = np.random.normal(loc=50, scale=10, size=30)
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

# 2. Statistiques descriptives
print("\nStatistiques descriptives :")
print(f"Moyenne avant formation : {np.mean(productivity_before):.2f}")
print(f"Moyenne après formation : {np.mean(productivity_after):.2f}")
print(f"Écart-type avant : {np.std(productivity_before):.2f}")
print(f"Écart-type après : {np.std(productivity_after):.2f}")

# 3. Test d'hypothèse (t-test apparié)
t_stat, p_value = stats.ttest_rel(productivity_after, productivity_before)

print("\nRésultats du test d'hypothèse :")
print(f"Statistique t : {t_stat:.2f}")
print(f"Valeur p : {p_value:.2f}")

# 4. Interprétation
print("\nInterprétation :")
if p_value < 0.05:
    print("La formation a eu un impact significatif sur la productivité (p < 0.05)")
else:
    print("Pas d'impact significatif de la formation (p > 0.05)")

# 5. Différence moyenne
difference = np.mean(productivity_after - productivity_before)
print(f"\nAugmentation moyenne de la productivité : {difference:.2f} points")
```

```python
print("EXERCICE 7 : COMPARAISON DE TABLEAUX")

# 1. Création de deux tableaux
np.random.seed(42)  # Pour la reproductibilité
tableau1 = np.random.randint(0, 100, 10)
tableau2 = np.random.randint(0, 100, 10)

# 2. Affichage des tableaux
print("\nTableau 1 :")
print(tableau1)
print("\nTableau 2 :")
print(tableau2)

# 3. Comparaison élément par élément
comparaison = tableau1 > tableau2

# 4. Affichage des résultats
print("\nRésultat de la comparaison (True si tableau1 > tableau2) :")
print(comparaison)

# 5. Statistiques sur la comparaison
print(f"\nNombre d'éléments plus grands dans le tableau 1 : {np.sum(comparaison)}")
print(f"Pourcentage d'éléments plus grands : {(np.sum(comparaison)/len(comparaison)*100):.2f}%")
```

```python
print("EXERCICE 8 : MANIPULATION DE SÉRIES TEMPORELLES")

# 1. Génération des données pour 2023
dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
np.random.seed(42)
valeurs = np.random.normal(loc=100, scale=10, size=len(dates))

# Création de la série temporelle
serie = pd.Series(valeurs, index=dates)

# 2. Découpage par trimestre
print("\nMoyennes trimestrielles :")

# Premier trimestre (Janvier-Mars)
q1 = serie['2023-01-01':'2023-03-31']
print(f"\nTrimestre 1 (Jan-Mar) : {q1.mean():.2f}")

# Deuxième trimestre (Avril-Juin)
q2 = serie['2023-04-01':'2023-06-30']
print(f"Trimestre 2 (Avr-Jun) : {q2.mean():.2f}")

# Troisième trimestre (Juillet-Septembre)
q3 = serie['2023-07-01':'2023-09-30']
print(f"Trimestre 3 (Jul-Sep) : {q3.mean():.2f}")

# Quatrième trimestre (Octobre-Décembre)
q4 = serie['2023-10-01':'2023-12-31']
print(f"Trimestre 4 (Oct-Dec) : {q4.mean():.2f}")

# 3. Statistiques par trimestre
print("\nNombre de jours par trimestre :")
print(f"T1 : {len(q1)} jours")
print(f"T2 : {len(q2)} jours")
print(f"T3 : {len(q3)} jours")
print(f"T4 : {len(q4)} jours")
```

```python
print("EXERCICE 9 : CONVERSION DE DONNÉES")

# 1. Création d'un tableau NumPy
np.random.seed(42)
tableau_numpy = np.random.randint(0, 100, size=(5, 3))
print("\nTableau NumPy original :")
print(tableau_numpy)

# 2. Conversion NumPy vers Pandas
df = pd.DataFrame(tableau_numpy, 
                 columns=['A', 'B', 'C'],
                 index=['Ligne 1', 'Ligne 2', 'Ligne 3', 'Ligne 4', 'Ligne 5'])
print("\nDataFrame Pandas converti :")
print(df)

# 3. Conversion Pandas vers NumPy
nouveau_tableau = df.to_numpy()
print("\nRetour vers tableau NumPy :")
print(nouveau_tableau)

# 4. Vérification des dimensions
print("\nDimensions :")
print(f"Tableau NumPy original : {tableau_numpy.shape}")
print(f"DataFrame Pandas : {df.shape}")
print(f"Nouveau tableau NumPy : {nouveau_tableau.shape}")
```
