```python
import numpy as np

# Création de la matrice des températures (7 jours, 4 villes)
temperature_data = np.array([
    [27, 24, 23, 22],  # Jour 1
    [20, 18, 23, 29],  # Jour 2
    [15, 20, 27, 23],  # Jour 3
    [18, 17, 25, 16],  # Jour 4
    [26, 19, 16, 20],  # Jour 5
    [18, 22, 21, 24],  # Jour 6
    [22, 21, 22, 28]   # Jour 7
])

# 1. Calcul de la température moyenne pour chaque ville
avg_temps = np.mean(temperature_data, axis=0)
print("Températures moyennes par ville:")
for i, avg in enumerate(avg_temps, 1):
    print(f"Ville {i}: {avg:.1f}°C")

# 2. Analyse des températures maximales et minimales
max_temps = np.max(temperature_data, axis=0)
min_temps = np.min(temperature_data, axis=0)
print("\nTempératures maximales et minimales:")
for i in range(4):
    print(f"Ville {i+1}:")
    print(f"  Maximum: {max_temps[i]}°C")
    print(f"  Minimum: {min_temps[i]}°C")
    print(f"  Amplitude thermique: {max_temps[i] - min_temps[i]}°C")

# 3. Comparaison entre le premier et le dernier jour
first_day = temperature_data[0]
last_day = temperature_data[-1]
temp_difference = last_day - first_day

print("\nComparaison Jour 1 vs Jour 7:")
for i in range(4):
    print(f"Ville {i+1}:")
    print(f"  Jour 1: {first_day[i]}°C")
    print(f"  Jour 7: {last_day[i]}°C")
    print(f"  Différence: {temp_difference[i]}°C")

# Analyses supplémentaires intéressantes
print("\nStatistiques supplémentaires:")

# Jour le plus chaud en moyenne
daily_averages = np.mean(temperature_data, axis=1)
hottest_day = np.argmax(daily_averages) + 1
print(f"Jour le plus chaud en moyenne: Jour {hottest_day} ({daily_averages[hottest_day-1]:.1f}°C)")

# Ville avec la plus grande variation de température
temp_variations = max_temps - min_temps
most_variable_city = np.argmax(temp_variations) + 1
print(f"Ville avec la plus grande variation de température: Ville {most_variable_city} ({temp_variations[most_variable_city-1]}°C)")
```
