```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.stats import ttest_ind

print("ANALYSE COMPARATIVE DES LONGUEURS DES PÉTALES D'IRIS")

# 1. Chargement et préparation des données
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Extraction des longueurs de pétales
setosa_petal_length = iris_df[iris_df['species'] == 'setosa']['petal length (cm)']
versicolor_petal_length = iris_df[iris_df['species'] == 'versicolor']['petal length (cm)']

# 2. Statistiques descriptives
print("\nStatistiques descriptives :")
print("\nIris Setosa :")
print(f"Moyenne : {setosa_petal_length.mean():.2f} cm")
print(f"Écart-type : {setosa_petal_length.std():.2f} cm")

print("\nIris Versicolor :")
print(f"Moyenne : {versicolor_petal_length.mean():.2f} cm")
print(f"Écart-type : {versicolor_petal_length.std():.2f} cm")

# 3. Test T
t_stat, p_val = ttest_ind(setosa_petal_length, versicolor_petal_length)

print("\nRésultats du Test T :")
print(f"Statistique T : {t_stat:.2f}")
print(f"Valeur P : {p_val:.2e}")

# 4. Visualisation
plt.figure(figsize=(12, 6))
sns.histplot(setosa_petal_length, color="skyblue", label="Iris Setosa", kde=True)
sns.histplot(versicolor_petal_length, color="red", label="Iris Versicolor", kde=True)

plt.title(f"Comparaison des Longueurs de Pétales\nT-statistique: {t_stat:.2f}, P-value: {p_val:.2e}")
plt.xlabel("Longueur des Pétales (cm)")
plt.ylabel("Fréquence")
plt.legend()
plt.grid(True)
plt.show()

# 5. Interprétation des résultats
print("\nInterprétation des résultats :")
print("1. Différence des moyennes :")
print(f"- La différence moyenne est de {abs(setosa_petal_length.mean() - versicolor_petal_length.mean()):.2f} cm")

print("\n2. Signification statistique :")
if p_val < 0.05:
    print("- La différence est statistiquement significative (p < 0.05)")
    print("- Nous pouvons rejeter l'hypothèse nulle")
else:
    print("- La différence n'est pas statistiquement significative (p > 0.05)")
    print("- Nous ne pouvons pas rejeter l'hypothèse nulle")

print("\n3. Analyse visuelle :")
print("- Les distributions montrent une séparation claire entre les deux espèces")
print("- Il y a peu de chevauchement entre les distributions")
```
