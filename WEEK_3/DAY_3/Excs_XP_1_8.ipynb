```python
print("Exercice 2 : Statistiques Descriptives")
print("-" * 50)

from scipy import stats
import numpy as np

# Dataset
data = [12, 15, 13, 12, 18, 20, 22, 21]

# Calcul des statistiques descriptives
mean = round(np.mean(data), 2)
median = round(np.median(data), 2)
variance = round(np.var(data), 2)
std_dev = round(np.std(data), 2)

# Affichage des résultats
print("\nStatistiques descriptives :")
print(f"Moyenne : {mean}")
print(f"Médiane : {median}")
print(f"Variance : {variance}")
print(f"Écart-type : {std_dev}")


```

```python
print("Exercice 3 : Comprendre les Distributions")
print("-" * 50)

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Générer une distribution normale
x = np.linspace(20, 80, 100)  # Points de 20 à 80
y = norm.pdf(x, 50, 10)  # Distribution normale (moyenne=50, écart-type=10)

# Création du graphique
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b-', label='Distribution Normale')
plt.title('Distribution Normale (μ=50, σ=10)')
plt.xlabel('Valeur')
plt.ylabel('Densité de probabilité')
plt.grid(True)
plt.legend()
plt.show()

# Statistiques de la distribution
print("\nStatistiques de la distribution :")
print(f"Moyenne théorique : {50.00}")
print(f"Écart-type théorique : {10.00}")
```

```python
print("Exercice 4 : Application du Test T")
print("-" * 50)

# Génération des données
np.random.seed(42)  # Pour reproductibilité
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

# Réalisation du test T
t_stat, p_value = stats.ttest_ind(data1, data2)

# Affichage des résultats
print("\nRésultats du Test T :")
print(f"Statistique T : {round(t_stat, 2)}")
print(f"Valeur P : {round(p_value, 2)}")

# Interprétation
print("\nInterprétation :")
if p_value < 0.05:
    print("Les deux ensembles sont significativement différents (p < 0.05)")
else:
    print("Pas de différence significative (p > 0.05)")
```

```python
print("Exercice 5 : Analyse de Régression Linéaire")
print("-" * 50)

# Données
house_sizes = np.array([50, 70, 80, 100, 120])
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

# Calcul de la régression linéaire
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

# Arrondir les résultats à 2 décimales
slope = round(slope, 2)
intercept = round(intercept, 2)

# Prédiction pour une maison de 90m²
prediction_90 = round(slope * 90 + intercept, 2)

print("\nRésultats de la régression :")
print(f"Pente : {slope} €/m²")
print(f"Ordonnée à l'origine : {intercept} €")
print(f"\nPrédiction pour 90m² : {prediction_90} €")

print("\nInterprétation :")
print(f"Pour chaque mètre carré supplémentaire, le prix augmente de {slope} €")
```

```python
print("Exercice 6 : Comprendre l'ANOVA")
print("-" * 50)

# Données
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

# Réalisation de l'ANOVA
f_stat, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

# Arrondir les résultats
f_stat = round(f_stat, 2)
p_value = round(p_value, 2)

print("\nRésultats de l'ANOVA :")
print(f"Valeur F : {f_stat}")
print(f"Valeur P : {p_value}")

print("\nInterprétation :")
if p_value < 0.05:
    print("Les fertilisants ont des effets significativement différents (p < 0.05)")
else:
    print("Pas de différence significative entre les fertilisants (p > 0.05)")

print("\nNote :")
print("Si p > 0.05, cela signifierait que les différences observées")
print("entre les fertilisants pourraient être dues au hasard.")
```

```python
print("Exercice 7 : Distributions de Probabilité")
print("-" * 50)

# Paramètres de la distribution binomiale
n = 10  # nombre de lancers
p = 0.5  # probabilité de succès (pile)
k = 5    # nombre de succès souhaité

# Calcul des probabilités
proba_exact = round(stats.binom.pmf(k, n, p), 2)  # Probabilité d'avoir exactement 5 piles
proba_moins = round(stats.binom.cdf(k, n, p), 2)  # Probabilité d'avoir 5 piles ou moins
proba_plus = round(1 - stats.binom.cdf(k-1, n, p), 2)  # Probabilité d'avoir 5 piles ou plus

print("\nProbabilités pour 10 lancers de pièce :")
print(f"Exactement 5 piles : {proba_exact}")
print(f"5 piles ou moins : {proba_moins}")
print(f"5 piles ou plus : {proba_plus}")

print("\nNote : La somme des probabilités de tous les événements possibles est 1")
```

```python
print("Exercice 8 : Coefficients de Corrélation")
print("-" * 50)

import pandas as pd
from scipy import stats

# Création du dataset
data = pd.DataFrame({
    'age': [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})

# Calcul des corrélations
pearson_coef, pearson_p = stats.pearsonr(data['age'], data['income'])
spearman_coef, spearman_p = stats.spearmanr(data['age'], data['income'])

# Arrondir les résultats
pearson_coef = round(pearson_coef, 2)
spearman_coef = round(spearman_coef, 2)

print("\nRésultats des corrélations :")
print(f"Coefficient de Pearson : {pearson_coef}")
print(f"Coefficient de Spearman : {spearman_coef}")

print("\nInterprétation :")
if abs(pearson_coef) > 0.7:
    print("Forte corrélation linéaire entre l'âge et le revenu")
elif abs(pearson_coef) > 0.3:
    print("Corrélation modérée entre l'âge et le revenu")
else:
    print("Faible corrélation entre l'âge et le revenu")
```
