```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

print("ANALYSE DE RÉGRESSION LINÉAIRE")

# 1. Création du dataset
x = np.array([2, 4, 6, 8, 10])
y = np.array([3, 5, 7, 9, 11])

# 2. Calcul de la régression linéaire
pente, ordonnee, r_value, p_value, std_err = stats.linregress(x, y)

print("\nRésultats de la régression :")
print(f"Pente (a) : {pente:.2f}")
print(f"Ordonnée à l'origine (b) : {ordonnee:.2f}")
print(f"Coefficient de détermination (R²) : {r_value**2:.4f}")
print(f"Erreur standard : {std_err:.4f}")
print(f"Valeur P : {p_value:.4e}")

# 3. Création de la ligne de régression
x_reg = np.linspace(x.min(), x.max(), 100)
y_reg = pente * x_reg + ordonnee

# 4. Visualisation
plt.figure(figsize=(10, 6))

# Graphique principal
plt.subplot(1, 2, 1)
plt.scatter(x, y, color='blue', label='Points de données')
plt.plot(x_reg, y_reg, color='red', label='Régression linéaire')
plt.title('Régression Linéaire')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.grid(True)
plt.legend()

# Graphique des résidus
residus = y - (pente * x + ordonnee)
plt.subplot(1, 2, 2)
plt.scatter(x, residus, color='green')
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Graphique des Résidus')
plt.xlabel('Variable X')
plt.ylabel('Résidus')
plt.grid(True)

plt.tight_layout()
plt.show()

# 5. Analyse détaillée
print("\nAnalyse détaillée :")

# Équation de la droite
print(f"\nÉquation de la droite : y = {pente:.2f}x + {ordonnee:.2f}")

# Interprétation de la pente
print("\nInterprétation de la pente :")
print(f"Pour chaque augmentation d'une unité de X, Y augmente de {pente:.2f} unités")

# Interprétation du R²
print("\nInterprétation du R² :")
print(f"Le modèle explique {(r_value**2 * 100):.1f}% de la variance dans les données")

# Analyse des résidus
print("\nAnalyse des résidus :")
print(f"Moyenne des résidus : {np.mean(residus):.4f}")
print(f"Écart-type des résidus : {np.std(residus):.4f}")

# Test de significativité
print("\nTest de significativité :")
if p_value < 0.05:
    print("La relation linéaire est statistiquement significative (p < 0.05)")
else:
    print("La relation linéaire n'est pas statistiquement significative (p > 0.05)")
```
