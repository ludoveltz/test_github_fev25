```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Création du dataset
data = pd.DataFrame({
    'Age': [25, 40, 30, 22, 35],
    'Monthly_Income': [3200, 5800, 4500, 2800, 5000],
    'Online_Purchase_Frequency': [15, 8, 12, 18, 10],
    'Hours_Social_Media': [30, 15, 25, 40, 20]
})

# Analyse des corrélations et visualisation
plt.figure(figsize=(12, 8))

# Matrice de corrélation
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Matrice de Corrélation des Variables')
plt.show()
```

Analyse Détaillée des Résultats
1. Coefficient de Corrélation de Pearson (0.75)
Relation Revenu Mensuel - Fréquence d'Achat en Ligne

Force de la relation : Corrélation positive forte (0.75)
Implications :
Les personnes à revenu plus élevé tendent à faire plus d'achats en ligne
56% (0.75²) de la variance dans la fréquence d'achat peut être expliquée par le revenu
Indication potentielle d'un pouvoir d'achat discrétionnaire
2. Corrélation de Spearman (-0.60)
Relation Âge - Temps sur les Réseaux Sociaux

Corrélation négative modérée (-0.60)
Hypothèses explicatives :
Les jeunes utilisent plus intensivement les réseaux sociaux
Différence générationnelle dans l'adoption du numérique
Possible contrainte de temps pour les personnes plus âgées
3. Analyse de la Covariance
Covariance Âge-Revenu (1500)
Covariance positive indiquant :
Tendance à l'augmentation du revenu avec l'âge
Possible effet de l'expérience professionnelle
Progression de carrière typique
Covariance Achats-Réseaux Sociaux (245)
Relation positive suggérant :
Influence possible des réseaux sociaux sur les comportements d'achat
Exposition accrue aux publicités en ligne
Effet potentiel du marketing d'influence
4. Corrélation vs Causalité
Points Importants à Considérer
Distinction Cruciale

La corrélation n'implique pas la causalité
D'autres facteurs peuvent influencer les relations observées
Exemples Concrets

Le revenu plus élevé ne cause pas directement plus d'achats
L'âge n'est pas la seule raison de l'utilisation des réseaux sociaux
