```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# 1. Gnration du dataset
np.random.seed(0)  # Pour la reproductibilit
x = np.arange(-1, 1, 0.1)  # 20 points entre -1 et 1
y = -x**2 + np.random.normal(0, 0.05, len(x))  # Fonction quadratique avec bruit

# 2. Visualisation des donnes
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Points avec bruit')
plt.plot(x, -x**2, 'r--', label='Fonction thorique (y = -x)')
plt.title('Relation avec Bruit Gaussien')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

# 3. Division en ensembles d'entranement et de test
x_train = x[:12]
y_train = y[:12]
x_test = x[12:]
y_test = y[12:]

# 4. Visualisation de la sparation train/test
plt.figure(figsize=(10, 6))
plt.scatter(x_train, y_train, color='blue', label='Donnes d\'entranement')
plt.scatter(x_test, y_test, color='red', label='Donnes de test')
plt.plot(x, -x**2, 'g--', label='Fonction thorique (y = -x)')
plt.title('Sparation Train/Test')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

# Affichage des dimensions des ensembles
print("Dimensions des ensembles :")
print(f"Entranement : {len(x_train)} points")
print(f"Test : {len(x_test)} points")

# Calcul de l'erreur quadratique moyenne par rapport  la fonction thorique
mse_train = mean_squared_error(y_train, -x_train**2)
mse_test = mean_squared_error(y_test, -x_test**2)

print("\nErreur quadratique moyenne :")
print(f"Sur l'ensemble d'entranement : {mse_train:.5f}")
print(f"Sur l'ensemble de test : {mse_test:.5f}")
```
