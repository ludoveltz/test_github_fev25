```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd  # Ajout de l'import manquant

def creer_diagramme_reseau():
    # Configuration du plot
    plt.figure(figsize=(12, 8))
    
    # Positions des neurones
    input_pos = np.array([[0, i] for i in range(3)])
    hidden_pos = np.array([[2, i] for i in range(4)])
    output_pos = np.array([[4, i] for i in range(2)])
    
    # Dessin des connexions
    for i in input_pos:
        for h in hidden_pos:
            plt.plot([i[0], h[0]], [i[1], h[1]], 'gray', alpha=0.5)
    
    for h in hidden_pos:
        for o in output_pos:
            plt.plot([h[0], o[0]], [h[1], o[1]], 'gray', alpha=0.5)
    
    # Dessin des neurones
    plt.scatter(input_pos[:,0], input_pos[:,1], c='lightblue', s=400, label='Couche d\'entre')
    plt.scatter(hidden_pos[:,0], hidden_pos[:,1], c='lightgreen', s=400, label='Couche cache')
    plt.scatter(output_pos[:,0], output_pos[:,1], c='salmon', s=400, label='Couche de sortie')
    
    # Annotations pour le dataset Mobile Price
    plt.text(-0.3, 2, 'RAM', fontsize=10)
    plt.text(-0.3, 1, 'Battery', fontsize=10)
    plt.text(-0.3, 0, 'Clock_speed', fontsize=10)
    
    plt.text(4.2, 1, 'Price High', fontsize=10)
    plt.text(4.2, 0, 'Price Low', fontsize=10)
    
    # Annotations techniques
    plt.annotate('w', xy=(0.5, 2.5), xytext=(0.5, 2.5))
    plt.annotate('ReLU', xy=(2, 3.5), xytext=(2, 3.5))
    plt.annotate('b', xy=(2, -0.5), xytext=(2, -0.5))
    
    # Lgende et labels
    plt.legend(loc='upper right')
    plt.title('Architecture du Rseau de Neurones pour Classification de Prix Mobile')
    plt.axis('off')
    plt.grid(False)
    
    plt.show()

# Affichage du diagramme
creer_diagramme_reseau()

# Description des composants
composants = pd.DataFrame({
    'Composant': ['Neurones', 'Poids (w)', 'Biais (b)', 'Fonction d\'activation', 'Couches'],
    'Description': [
        'Units de traitement qui reoivent, transforment et transmettent l\'information',
        'Paramtres ajustables qui dterminent l\'importance de chaque connexion',
        'Valeurs ajoutes qui permettent de dcaler la fonction d\'activation',
        'Fonction non-linaire (ReLU) qui dtermine l\'activation du neurone',
        'Groupes de neurones organiss en couches successives'
    ]
})

display(composants)
```

Le flux d'information dans un rseau de neurones artificiel suit un processus squentiel de propagation avant (forward propagation). Les donnes d'entre sont d'abord reues par la couche d'entre, o chaque neurone reprsente une caractristique du problme (par exemple, pour la prdiction du temps de montage vido : dure des rushes, complexit, nombre de squences). Ces valeurs sont ensuite pondres et transmises  la couche cache, o chaque neurone applique une fonction d'activation (comme ReLU)  la somme pondre de ses entres plus un biais. Enfin, la couche de sortie reoit les activations de la couche cache et produit les prdictions finales (par exemple, le temps de montage estim).
