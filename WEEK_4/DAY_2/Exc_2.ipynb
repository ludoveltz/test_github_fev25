```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd  # Ajout de l'import manquant

def creer_diagramme_reseau():
    # Configuration du plot
    plt.figure(figsize=(12, 8))
    
    # Positions des neurones
    input_pos = np.array([[0, i] for i in range(3)])
    hidden_pos = np.array([[2, i] for i in range(4)])
    output_pos = np.array([[4, i] for i in range(2)])
    
    # Dessin des connexions
    for i in input_pos:
        for h in hidden_pos:
            plt.plot([i[0], h[0]], [i[1], h[1]], 'gray', alpha=0.5)
    
    for h in hidden_pos:
        for o in output_pos:
            plt.plot([h[0], o[0]], [h[1], o[1]], 'gray', alpha=0.5)
    
    # Dessin des neurones
    plt.scatter(input_pos[:,0], input_pos[:,1], c='lightblue', s=400, label='Couche d\'entrée')
    plt.scatter(hidden_pos[:,0], hidden_pos[:,1], c='lightgreen', s=400, label='Couche cachée')
    plt.scatter(output_pos[:,0], output_pos[:,1], c='salmon', s=400, label='Couche de sortie')
    
    # Annotations pour le dataset Mobile Price
    plt.text(-0.3, 2, 'RAM', fontsize=10)
    plt.text(-0.3, 1, 'Battery', fontsize=10)
    plt.text(-0.3, 0, 'Clock_speed', fontsize=10)
    
    plt.text(4.2, 1, 'Price High', fontsize=10)
    plt.text(4.2, 0, 'Price Low', fontsize=10)
    
    # Annotations techniques
    plt.annotate('w₁₁', xy=(0.5, 2.5), xytext=(0.5, 2.5))
    plt.annotate('ReLU', xy=(2, 3.5), xytext=(2, 3.5))
    plt.annotate('b₁', xy=(2, -0.5), xytext=(2, -0.5))
    
    # Légende et labels
    plt.legend(loc='upper right')
    plt.title('Architecture du Réseau de Neurones pour Classification de Prix Mobile')
    plt.axis('off')
    plt.grid(False)
    
    plt.show()

# Affichage du diagramme
creer_diagramme_reseau()

# Description des composants
composants = pd.DataFrame({
    'Composant': ['Neurones', 'Poids (w)', 'Biais (b)', 'Fonction d\'activation', 'Couches'],
    'Description': [
        'Unités de traitement qui reçoivent, transforment et transmettent l\'information',
        'Paramètres ajustables qui déterminent l\'importance de chaque connexion',
        'Valeurs ajoutées qui permettent de décaler la fonction d\'activation',
        'Fonction non-linéaire (ReLU) qui détermine l\'activation du neurone',
        'Groupes de neurones organisés en couches successives'
    ]
})

display(composants)
```

Le flux d'information dans un réseau de neurones artificiel suit un processus séquentiel de propagation avant (forward propagation). Les données d'entrée sont d'abord reçues par la couche d'entrée, où chaque neurone représente une caractéristique du problème (par exemple, pour la prédiction du temps de montage vidéo : durée des rushes, complexité, nombre de séquences). Ces valeurs sont ensuite pondérées et transmises à la couche cachée, où chaque neurone applique une fonction d'activation (comme ReLU) à la somme pondérée de ses entrées plus un biais. Enfin, la couche de sortie reçoit les activations de la couche cachée et produit les prédictions finales (par exemple, le temps de montage estimé).
