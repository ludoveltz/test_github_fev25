```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Implémentation des fonctions d'activation
def step_function(x):
    return np.where(x >= 0, 1, 0)

def sigmoid_function(x):
    return 1 / (1 + np.exp(-x))

def relu_function(x):
    return np.maximum(0, x)

# 2. Création du graphique
x = np.linspace(-10, 10, 1000)

plt.figure(figsize=(12, 6))
plt.plot(x, step_function(x), label='Step', linestyle='--')
plt.plot(x, sigmoid_function(x), label='Sigmoid')
plt.plot(x, relu_function(x), label='ReLU')

plt.title('Comparaison des Fonctions d\'Activation')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.show()

# 3. Comparaison des comportements
print("Analyse des comportements :")
print("\nStep Function :")
print("- Sortie binaire (0 ou 1)")
print("- Pas de transition douce")
print("- Utilisée historiquement dans les premiers perceptrons")

print("\nSigmoid Function :")
print("- Transition douce entre 0 et 1")
print("- Sortie bornée entre 0 et 1")
print("- Dérivable sur tout son domaine")

print("\nReLU Function :")
print("- Conserve les valeurs positives")
print("- Met à zéro les valeurs négatives")
print("- Calcul simple et efficace")

# 4. Réponses aux questions
print("\nQuestions et Réponses :")

print("\nPourquoi ReLU est couramment utilisée en deep learning ?")
print("- Calcul rapide et simple")
print("- Résout le problème de la disparition du gradient")
print("- Introduit de la non-linéarité sans saturation pour les valeurs positives")
print("- Permet une convergence plus rapide pendant l'entraînement")

print("\nPourquoi Sigmoid est adaptée pour la classification binaire ?")
print("- Sortie bornée entre 0 et 1")
print("- Peut être interprétée comme une probabilité")
print("- Transition douce permettant une bonne discrimination")
print("- Dérivable, facilitant l'apprentissage")

print("\nFaiblesses de la fonction Step ?")
print("- Non dérivable (impossible d'utiliser la descente de gradient)")
print("- Pas de nuance dans la sortie (uniquement 0 ou 1)")
print("- Ne permet pas d'apprentissage graduel")
print("- Transition brutale pouvant causer des instabilités")
```
