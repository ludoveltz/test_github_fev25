Exercice de Graphique Linéaire (y = x²)


```python
import matplotlib.pyplot as plt
import numpy as np

# Création des données pour x et y=x²
x = np.linspace(0, 10, 100)  # 100 points entre 0 et 10
y = x**2

# Création du graphique
plt.figure(figsize=(8, 6))  # Taille de la figure
plt.plot(x, y, 'b-', linewidth=2)  # 'b-' pour ligne bleue continue

# Ajout des labels et du titre
plt.title('Fonction y = x²')
plt.xlabel('Valeurs de x')
plt.ylabel('Valeurs de y')

# Ajout d'une grille
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

Exercice de Diagramme en Barres


```python
import matplotlib.pyplot as plt

# Données
produits = ['Produit A', 'Produit B', 'Produit C', 'Produit D', 'Produit E']
ventes = [120, 98, 145, 87, 156]
couleurs = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FF99CC']
motifs = ['/', '\\', 'x', 'o', '*']

# Création du diagramme en barres
plt.figure(figsize=(10, 6))
barres = plt.bar(produits, ventes, color=couleurs)

# Personnalisation des barres avec des motifs
for barre, motif in zip(barres, motifs):
    barre.set_hatch(motif)

# Ajout des labels et du titre
plt.title('Ventes par Produit')
plt.xlabel('Produits')
plt.ylabel('Nombre de Ventes')

# Ajout des valeurs au-dessus des barres
for barre in barres:
    hauteur = barre.get_height()
    plt.text(barre.get_x() + barre.get_width()/2., hauteur,
             f'{int(hauteur)}',
             ha='center', va='bottom')

plt.show()
```

Exercice d'Histogramme


```python
import matplotlib.pyplot as plt
import numpy as np

# Génération des données
donnees = np.random.normal(loc=0, scale=1, size=500)

# Création de l'histogramme avec différentes configurations
plt.figure(figsize=(12, 4))

# Trois histogrammes avec différents nombres de bins
plt.subplot(131)
plt.hist(donnees, bins=10, color='skyblue', edgecolor='black')
plt.title('10 intervalles')

plt.subplot(132)
plt.hist(donnees, bins=20, color='lightgreen', edgecolor='black')
plt.title('20 intervalles')

plt.subplot(133)
plt.hist(donnees, bins=30, color='salmon', edgecolor='black')
plt.title('30 intervalles')

plt.tight_layout()
plt.show()
```

Exercice de Diagramme Circulaire


```python
import matplotlib.pyplot as plt

# Données des activités quotidiennes
activites = ['Travail', 'Sommeil', 'Loisirs', 'Sport', 'Repas']
temps = [8, 7, 4, 2, 3]  # en heures
couleurs = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FF99CC']

# Création du diagramme circulaire
plt.figure(figsize=(10, 8))
explode = (0.1, 0, 0, 0, 0)  # Met en évidence le travail

plt.pie(temps, 
        explode=explode,
        labels=activites,
        colors=couleurs,
        autopct='%1.1f%%',
        shadow=True,
        startangle=90)

plt.title('Répartition du Temps Quotidien')
plt.axis('equal')  # Assure que le cercle est rond

plt.show()
```
