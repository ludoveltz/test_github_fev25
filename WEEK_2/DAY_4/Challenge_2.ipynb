```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import kagglehub

# Téléchargement du dataset depuis Kaggle
path = kagglehub.dataset_download("notepub/tips-dataset-for-beginners")

# Lecture du fichier CSV (ajustez le nom du fichier selon votre dataset)
tips = pd.read_csv(f"{path}/tips.csv")

# Création de l'histogramme personnalisé
plt.figure(figsize=(10, 6))
sns.histplot(data=tips, 
             x="total_bill",  # Assurez-vous que le nom de la colonne correspond à votre dataset
             bins=30,
             color="skyblue",
             edgecolor="black",
             kde=True)  # Ajout de la courbe de densité

# Personnalisation du graphique
plt.title("Distribution des Dépenses Clients")
plt.xlabel("Montant Total de l'Addition (\$)")
plt.ylabel("Nombre de Clients")

plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Chargement du dataset Iris
iris_path = "./WEEK_2/DAY_2/DATASET/Iris_dataset.csv"
iris = pd.read_csv(iris_path)

# Renommage des colonnes
iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'Species']

# 2. Pairplot avec le dataset Iris
sns.pairplot(iris,
             hue="Species",  # Notez le 'S' majuscule
             palette="Set2",
             markers=["o", "s", "D"],
             diag_kind="hist",
             plot_kws={"alpha": 0.6})
plt.suptitle("Visualisation des Relations entre Variables par Espèce", y=1.02)
plt.show()

# 3. Heatmap des corrélations avec Iris
plt.figure(figsize=(8, 6))
# On exclut la colonne Species car c'est une variable catégorielle
numeric_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
corr_matrix = iris[numeric_columns].corr()
sns.heatmap(corr_matrix,
            annot=True,
            cmap="coolwarm",
            center=0,
            vmin=-1,
            vmax=1,
            square=True)
plt.title("Matrice de Corrélation des Variables de l'Iris")
plt.show()

# 4. FacetGrid avec le dataset Titanic
titanic_path = "./WEEK_2/DAY_3/DATASET/Titanic-Dataset.csv"
titanic = pd.read_csv(titanic_path)
g = sns.FacetGrid(titanic,
                  col="Pclass",
                  row="Sex",
                  height=4,
                  aspect=1.5)
g.map_dataframe(sns.scatterplot,
                x="Age",
                y="Survived",
                alpha=0.6)
g.fig.suptitle("Taux de Survie par Âge, Classe et Genre", y=1.02)
for ax in g.axes.flat:
    ax.set_xlabel("Âge")
    ax.set_ylabel("Survie (0=Non, 1=Oui)")
plt.show()


```
