```python
import matplotlib.pyplot as plt

# Données
jours = ['Lundi', 'Mardi', 'Mercredi', 'Jeudi', 'Vendredi', 'Samedi', 'Dimanche']
temperatures = [72, 74, 76, 80, 82, 78, 75]

# Création du graphique
plt.figure(figsize=(10, 6))  # Définir la taille du graphique

# Tracer la ligne avec des marqueurs
plt.plot(jours, temperatures, marker='o', linestyle='-', linewidth=2, markersize=8)

# Personnalisation du graphique
plt.title('Variation de la Température sur une Semaine', fontsize=14, pad=15)
plt.xlabel('Jour', fontsize=12)
plt.ylabel('Température (°F)', fontsize=12)

# Ajout d'une grille
plt.grid(True, linestyle='--', alpha=0.7)

# Rotation des étiquettes des jours pour une meilleure lisibilité
plt.xticks(rotation=45)

# Ajout des valeurs de température sur chaque point
for i, temp in enumerate(temperatures):
    plt.text(i, temp + 0.5, f'{temp}°F', ha='center')

# Ajustement automatique de la mise en page
plt.tight_layout()

# Affichage du graphique
plt.show()

```
