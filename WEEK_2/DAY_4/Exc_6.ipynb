```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Chargement des données
df = pd.read_csv('./WEEK_2/DAY_4/DATASET/Student Mental health.csv')

# Vérification des colonnes
print("Aperçu des colonnes :")
print(df.columns)
print("\nPremières lignes :")
print(df.head())
```

```python
# Conversion des réponses "Yes/No" en valeurs numériques (1/0)
df['Panic_Attack_Numeric'] = (df['Do you have Panic attack?'] == 'Yes').astype(int)

# Création du scatter plot
plt.figure(figsize=(10, 6))

# Ajout de "jitter" pour mieux voir la distribution des points
sns.scatterplot(
    data=df,
    x='Age',
    y='Panic_Attack_Numeric',
    alpha=0.5,  # Transparence des points
    s=100,      # Taille des points
    color='purple'  # Couleur des points
)

# Personnalisation du graphique
plt.title('Relation entre l\'Âge et les Attaques de Panique', fontsize=14, pad=15)
plt.xlabel('Âge', fontsize=12)
plt.ylabel('Attaques de Panique (0=Non, 1=Oui)', fontsize=12)

# Ajout d'une grille pour une meilleure lisibilité
plt.grid(True, alpha=0.3)

# Ajustement des limites de l'axe y pour mieux voir les points
plt.ylim(-0.2, 1.2)

# Ajustement automatique de la mise en page
plt.tight_layout()

# Affichage des statistiques
print("\nStatistiques descriptives :")
print("\nNombre d'étudiants ayant des attaques de panique par groupe d'âge :")
print(df.groupby('Age')['Do you have Panic attack?'].value_counts().unstack())

print("\nPourcentage d'étudiants ayant des attaques de panique :")
print(f"{(df['Panic_Attack_Numeric'].mean() * 100).round(2)}%")

# Affichage du graphique
plt.show()
```
