```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Chargement des données
df = pd.read_csv('./WEEK_2/DAY_4/DATASET/MentalHealthSurvey.csv')

# Vérification des données
print("Aperçu des colonnes :")
print(df.columns)
print("\nPremières lignes :")
print(df.head())
```

```python
# Création du graphique
plt.figure(figsize=(10, 6))

# Création du bar plot avec Seaborn
sns.barplot(
    data=df,
    x='gender',
    y='anxiety',
    hue='gender',  
    errorbar=('ci', 95),  
    legend=False  # Cache la légende car redondante avec l'axe x
)

# Personnalisation du graphique
plt.title('Niveaux d\'Anxiété par Genre', fontsize=14, pad=15)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Niveau d\'Anxiété (1-5)', fontsize=12)

# Ajout d'une grille pour une meilleure lisibilité
plt.grid(True, alpha=0.3, axis='y')

# Ajustement automatique de la mise en page
plt.tight_layout()

# Affichage des statistiques
print("\nMoyenne d'anxiété par genre :")
print(df.groupby('gender')['anxiety'].mean().round(2))
print("\nNombre de personnes par genre :")
print(df['gender'].value_counts())

# Affichage du graphique
plt.show()

```
