```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Chargement des données
df = pd.read_csv('./WEEK_2/DAY_4/DATASET/placement-dataset.csv')

# Création de l'histogramme avec Seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data=df, 
             x='cgpa',
             color='skyblue',  # Couleur personnalisée
             bins=20,          # Nombre de bins pour l'histogramme
             kde=True)         # Ajoute une courbe de densité

# Personnalisation du graphique
plt.title('Distribution des CGPA des Étudiants', fontsize=14, pad=15)
plt.xlabel('CGPA', fontsize=12)
plt.ylabel('Nombre d\'étudiants', fontsize=12)

# Ajout d'une grille pour une meilleure lisibilité
plt.grid(True, alpha=0.3)

# Ajustement automatique de la mise en page
plt.tight_layout()

# Affichage du graphique
plt.show()
```
