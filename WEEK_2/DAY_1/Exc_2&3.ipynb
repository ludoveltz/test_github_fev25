```python

# Fonction pour catégoriser les colonnes
def categorize_columns(dataframe, name):
    print(f"\n=== ANALYSE DES TYPES DE DONNÉES : {name} ===\n")
    
    for column in dataframe.columns:
        dtype = dataframe[column].dtype
        unique_values = dataframe[column].nunique()
        
        # Détermine si la colonne est quantitative ou qualitative
        if pd.api.types.is_numeric_dtype(dataframe[column]):
            if len(dataframe[column].unique()) < 10:
                category = "Qualitative (Discrète)"
            else:
                category = "Quantitative"
        else:
            category = "Qualitative (Catégorielle)"
            
        print(f"Colonne : {column}")
        print(f"Type : {category}")
        print(f"Type de données Python : {dtype}")
        print(f"Nombre de valeurs uniques : {unique_values}")
        print("-" * 50)

# Analyse pour chaque dataset
print("1. DATASET CARTES DE CRÉDIT")
categorize_columns(credit_df, "Cartes de Crédit")

print("\n2. DATASET SANTÉ MENTALE")
categorize_columns(mental_health_df, "Santé Mentale")

print("\n3. DATASET SOMMEIL")
categorize_columns(sleep_df, "Sommeil")
```

Explication du raisonnement :

Pour le Dataset Cartes de Crédit :
Variables Quantitatives :
Les montants, âges, scores numériques
Raison : Représentent des mesures continues ou discrètes avec beaucoup de valeurs possibles

Variables Qualitatives :
Statut d'approbation, type de carte
Raison : Représentent des catégories ou des attributs


Pour le Dataset Santé Mentale :
Variables Quantitatives :
Taux de dépression, années
Raison : Mesures numériques continues

Variables Qualitatives :
Pays, régions
Raison : Catégories nominales


Pour le Dataset Sommeil :
Variables Quantitatives :
Heures de sommeil, temps
Raison : Mesures continues de durée

Variables Qualitatives :
Catégories démographiques
Raison : Groupes ou caractéristiques distinctes
