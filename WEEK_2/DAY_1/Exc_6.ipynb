```python
# Import des bibliothèques nécessaires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Chargement du dataset
sleep_df = pd.read_csv('WEEK_2/DAY_1/DATASET/Time Americans Spend Sleeping.csv')
```

```python
# 1. Exploration initiale du dataset
print("=== APERÇU DU DATASET ===")
print("\nPremières lignes :")
print(sleep_df.head())

print("\nInformations sur les colonnes :")
print(sleep_df.info())
```

```python
# 2. Analyse des colonnes intéressantes pour différents types d'analyses
print("\n=== ANALYSE DES COLONNES POTENTIELLES ===")

# Liste des colonnes et leur potentiel d'analyse
analysis_potential = {
    "Trend Analysis": {
        "columns": ["Year", "Average Hours of Sleep"],
        "reason": "Can show how sleep patterns have changed over time"
    },
    "Group Comparison": {
        "columns": ["Age Group", "Gender", "Average Hours of Sleep"],
        "reason": "Can compare sleep patterns across different demographics"
    },
    "Correlation Analysis": {
        "columns": ["Work Hours", "Average Hours of Sleep"],
        "reason": "Can explore relationship between work and sleep"
    }
}

# Affichage des analyses potentielles
for analysis_type, details in analysis_potential.items():
    print(f"\n{analysis_type}:")
    print(f"Columns: {', '.join(details['columns'])}")
    print(f"Reason: {details['reason']}")
    
    # Statistiques descriptives pour ces colonnes
    try:
        print("\nDescriptive Statistics:")
        print(sleep_df[details['columns']].describe())
    except:
        print("Cannot generate statistics for these columns")
```

```python
# 3. Visualisations pour supporter l'analyse
print("\n=== VISUALISATIONS EXPLORATOIRES ===")
# a. Évolution temporelle du sommeil
plt.figure(figsize=(12, 6))
sns.lineplot(data=sleep_df, x='Year', y='Avg hrs per day sleeping')
plt.title('Evolution of Average Sleep Hours Over Time')
plt.xlabel('Year')
plt.ylabel('Average Hours of Sleep per Day')
plt.xticks(rotation=45)
plt.show()

# b. Distribution des heures de sommeil par groupe d'âge
plt.figure(figsize=(12, 6))
sns.boxplot(data=sleep_df, x='Age Group', y='Avg hrs per day sleeping')
plt.title('Sleep Hours Distribution by Age Group')
plt.xticks(rotation=45)
plt.show()

# c. Comparaison par genre
plt.figure(figsize=(10, 6))
sns.barplot(data=sleep_df, x='Sex', y='Avg hrs per day sleeping')
plt.title('Average Sleep Hours by Gender')
plt.show()

# d. Analyse statistique
print("\n=== ANALYSE STATISTIQUE ===")
print("\nStatistiques descriptives des heures de sommeil:")
print(sleep_df['Avg hrs per day sleeping'].describe())

print("\nMoyenne des heures de sommeil par groupe d'âge:")
print(sleep_df.groupby('Age Group')['Avg hrs per day sleeping'].mean().round(2))

print("\nMoyenne des heures de sommeil par type de jour:")
print(sleep_df.groupby('Type of Days')['Avg hrs per day sleeping'].mean().round(2))
```
