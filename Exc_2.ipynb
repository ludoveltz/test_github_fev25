```python
# 1. Import de la classe pipeline
from transformers import pipeline

# 2. Création du pipeline d'analyse de sentiment avec le modèle spécifié
sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)

# 3. Utilisation du texte d'exemple exact donné dans l'exercice
text = "I really enjoyed this movie! It was fantastic."

# 4. Prédiction du sentiment avec le pipeline
result = sentiment_analyzer(text)

# 5. Affichage du label et du score de confiance
print("\nTexte analysé :")
print(text)
print("\nRésultat de l'analyse :")
print(f"Sentiment : {result[0]['label']}")
print(f"Score de confiance : {result[0]['score']:.4f}")
```
